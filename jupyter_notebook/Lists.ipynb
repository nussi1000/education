```python
# import required modules
import requests, json

#You can create your Key here: https://openweathermap.org/api

# Enter your API key here
api_key = "ENTER KEY HERE"
base_url = "http://api.openweathermap.org/data/2.5/weather?"

# Give city name
#city_name = input("Enter city names (1 or more, split by ',') : ").split(',')
city_name = ['Bern', 'Zurich', 'New York', 'New Bern', 'Los Angeles', 'Moscow']

somelist =[]

for i in city_name:
	city = i
	# complete_url variable to store
	# complete url address
	complete_url = base_url + "appid=" + api_key + "&q=" + str(city)

	# get method of requests module
	# return response object
	response = requests.get(complete_url)

	# json method of response object
	# convert json format data into
	# python format data
	x = response.json()
	if x["cod"] != "404":
			

			# store the value of "main"
			# key in variable y
			y = x["main"]

			# store the value corresponding
			# to the "temp" key of y
			current_temperature = y["temp"]
			current_temperature_celsius = current_temperature -273.15 

			# store the value corresponding
			# to the "pressure" key of y
			current_pressure = y["pressure"]

			# store the value corresponding
			# to the "humidity" key of y
			current_humidity = y["humidity"]

			# get coordinates of the city
			coordinates = x["coord"]
			coordinates_lon = coordinates["lon"]
			coordinates_lat = coordinates["lat"]

			# store the value of "weather"
			# key in variable z
			z = x["weather"]

			# store the value corresponding
			# to the "description" key at
			# the 0th index of z
			weather_description = z[0]["description"]
	
	somelist.append(city+" / "+weather_description+" / "+str(round(current_temperature_celsius, 3)))
print(somelist)


city = 'Denpasar'
# complete_url variable to store
# complete url address
complete_url = base_url + "appid=" + api_key + "&q=" + str(city)

# get method of requests module
# return response object
response = requests.get(complete_url)

# json method of response object
# convert json format data into
# python format data
x = response.json()
if x["cod"] != "404":
			

		# store the value of "main"
		# key in variable y
	y = x["main"]

		# store the value corresponding
		# to the "temp" key of y
	current_temperature = y["temp"]
	current_temperature_celsius = current_temperature -273.15 

			# store the value corresponding
			# to the "pressure" key of y
	current_pressure = y["pressure"]

			# store the value corresponding
			# to the "humidity" key of y
	current_humidity = y["humidity"]

			# get coordinates of the city
	coordinates = x["coord"]
	coordinates_lon = coordinates["lon"]
	coordinates_lat = coordinates["lat"]

	# store the value of "weather"
	# key in variable z
	z = x["weather"]

	# store the value corresponding
	# to the "description" key at
	# the 0th index of z
	weather_description = z[0]["description"]

somelist.insert(3,city + " / " + weather_description + " / " + str(round(current_temperature_celsius, 3)))
print(somelist)

cutout=somelist.pop(2)
print(somelist)
print(cutout)

#somelist.index('New York')

elementamount=len(somelist)
print(elementamount)

somelist.sort()
print(somelist)
```
