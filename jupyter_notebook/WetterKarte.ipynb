```python
# import required modules
import requests, json, time, pandas

api_key

# base_url variable to store urls
base_url = "http://api.openweathermap.org/data/2.5/weather?"

while True:

	# Give city name
	city_name = input("Enter city names (1 or more, split by ',') : ").split(',')

	var = pandas.DataFrame()

	for i in city_name:
		city_name = i
		# complete_url variable to store
		# complete url address
		complete_url = base_url + "appid=" + api_key + "&q=" + city_name

		# get method of requests module
		# return response object
		response = requests.get(complete_url)

		# json method of response object
		# convert json format data into
		# python format data
		x = response.json()

		# Now x contains list of nested dictionaries
		# Check the value of "cod" key is equal to
		# "404", means city is found otherwise,
		# city is not found
		if x["cod"] != "404":
			

			# store the value of "main"
			# key in variable y
			y = x["main"]
			z = x["sys"]

			# store the value corresponding
			# to the "temp" key of y
			current_temperature = y["temp"]
			current_temperature_celsius = current_temperature -273.15 

			# store the value corresponding
			# to the "pressure" key of y
			current_pressure = y["pressure"]

			# store the value corresponding
			# to the "humidity" key of y
			current_humidity = y["humidity"]

			#Get country
			current_country = z["country"]

			# get coordinates of the city
			coordinates = x["coord"]
			coordinates_lon = coordinates["lon"]
			coordinates_lat = coordinates["lat"]

			# store the value of "weather"
			# key in variable z
			z = x["weather"]

			# store the value corresponding
			# to the "description" key at
			# the 0th index of z
			weather_description = z[0]["description"]

			# print following values
			
			dataset = {
			'City': [city_name],
			'Country': [current_country],
			'Temparature': [current_temperature_celsius],
			'Weather': [weather_description]
			}

			varfill = pandas.DataFrame(dataset)	
			
			var = var.append(varfill, ignore_index = True)
		
		else:
			print(" City " +city_name+ " not found ")

	print(var)

	response = input("Would you like to continue? (y/n) ")
	if response == 'n':
		print("Stopping Programm")
		time.sleep(1)
		print("Goodbye ;D")
		break
	elif response == 'y':
		print("Restarting Programm")
		time.sleep(0.5)

	else:
		print("Invalid input, please awnser with y/n.")
		time.sleep(3)
		print("Goodbye ;D")
		break


```

```python
# import required modules
import requests, json, time

# Enter your API key here
api_key 

# base_url variable to store url
base_url = "http://api.openweathermap.org/data/2.5/weather?"

while True:

	# Give city name
	city_name = input("Enter city names (1 or more, split by ',') : ").split(',')

	for i in city_name:
		print("---City/Village : "+i+" ---")
		city_name = i
		# complete_url variable to store
		# complete url address
		complete_url = base_url + "appid=" + api_key + "&q=" + city_name

		# get method of requests module
		# return response object
		response = requests.get(complete_url)

		# json method of response object
		# convert json format data into
		# python format data
		x = response.json()

		# Now x contains list of nested dictionaries
		# Check the value of "cod" key is equal to
		# "404", means city is found otherwise,
		# city is not found
		if x["cod"] != "404":
			

			# store the value of "main"
			# key in variable y
			y = x["main"]

			# store the value corresponding
			# to the "temp" key of y
			current_temperature = y["temp"]
			current_temperature_celsius = current_temperature -273.15 

			# store the value corresponding
			# to the "pressure" key of y
			current_pressure = y["pressure"]

			# store the value corresponding
			# to the "humidity" key of y
			current_humidity = y["humidity"]

			# get coordinates of the city
			coordinates = x["coord"]
			coordinates_lon = coordinates["lon"]
			coordinates_lat = coordinates["lat"]

			# store the value of "weather"
			# key in variable z
			z = x["weather"]

			# store the value corresponding
			# to the "description" key at
			# the 0th index of z
			weather_description = z[0]["description"]

			# print following values
			
			
			
			print(
				" Coordinates = Lat: " +
							str(coordinates_lat) + " Lon: " + str(coordinates_lon) +
				"\n Temperature (in celsius unit) = " +
							str(round(current_temperature_celsius, 2)) +
				"\n atmospheric pressure (in hPa unit) = " +
							str(current_pressure) +
				"\n humidity (in percentage) = " +
							str(current_humidity) +
				"\n description = " +
							str(weather_description) + 
				"\n " )
			

		else:
			print(" City Not Found ")

	response = input("Would you like to continue? (y/n) ")
	if response == 'n':
		print("Stopping Programm")
		time.sleep(0.5)
		print(".", end='\r')
		time.sleep(0.5)
		print("..", end='\r')
		time.sleep(0.5)
		print("...", end='\r')
		time.sleep(0.5)
		print("....", end='\r')
		time.sleep(0.5)
		print(".....", end='\r')
		time.sleep(1)
		print("Goodbye ;D")
		break
	elif response == 'y':
		print("Restarting Programm")
		time.sleep(0.5)
		print(".", end='\r')
		time.sleep(0.5)
		print("..", end='\r')
		time.sleep(0.5)
		print("...", end='\r')
		time.sleep(0.5)
		print("....", end='\r')
		time.sleep(0.5)
		print(".....", end='\r')
	else:
		print("Invalid input, please awnser with y/n.")
		time.sleep(5)
		print("Goodbye ;D")
		break


```
